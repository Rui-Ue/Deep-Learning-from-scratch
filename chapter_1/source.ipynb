# rpy2 でカーネル切り替え省略

１章は Python の使い方がほとんどなので，R とも比較しながらちょっと流し読み．  

本書では使ってないが，[rpy2 ライブラリ](https://github.com/rpy2) を使えば，マジックコマンドの利用で済みカーネル切り替え要らず．

```python
%load_ext rpy2.ipython
```

```python
# 例

print( 2**3 )              # Python コード
%R print( 2^3 )         # R コード

print( True and False )
%R print( TRUE && FALSE )

print( not False )
%R print( !FALSE )
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# np.array (書いてないが pd.Series も) では R ライクなベクトル演算できる．

x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print( x + y )
print( x - y )
print( x * y )  # 要素ごとの積で，内積ではない．
print( np.dot(x, y) )  # 内積
print( x / y )
print( x / 10 )
print( x + 9 )

%R x = c(1, 2, 3)
%R y = c(2, 4, 6)
%R print( x + y)
%R print( x - y )
%R print( x * y )
%R print( x %*% y )
%R print( x / y )
%R print( x / 10 )
%R print( x + 9 )
```

```python
# matrix (2d array)

print( np.array([[1, 2, 3], [4, 5, 6]]) )
%R print( matrix(c(1, 2, 3, 4, 5, 6), nrow=2) )
```

```python
# dataframe

df = pd.DataFrame({
    "Sepal_Length": [2, 1, 3, 4],
    "Sepal_Width":[3, 4, 1, 2]
})
print( df )

%R df = data.frame(Sepal_Length = c(2, 1, 3, 4), Sepal_Width = c(3, 4, 1, 2))
%R print(df)
```

```python
# 図

x = np.arange(0, 6, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()

%R x = seq(0, 6, 0.1)
%R y = sin(x)
%R plot(x, y)
```
